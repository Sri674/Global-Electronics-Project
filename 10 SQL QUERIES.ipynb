```python
mycursor=mydb.cursor(buffered=True)
```

```python
mycursor.execute("USE dataspark")
```

```python
!pip install tabulate
```

```python
            #TOTAL SALES FOR A SPECIFIC PERIOD

mycursor.execute("SELECT order_date,SUM(quantity * unit_price_usd) AS TotalSales FROM overall WHERE order_date BETWEEN '2019-12-1' AND '2021-02-20'GROUP BY order_date")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
                       #SALES BY PRODUCT CATEGORY

mycursor.execute("SELECT category,SUM(quantity * unit_price_usd) AS TotalSales FROM overall GROUP BY category ORDER BY TotalSales DESC")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))

```

```python
                         #TOP 5 CUSTOMERS BY TOTAL SPEND

mycursor.execute("SELECT name, SUM(quantity * unit_price_usd) AS Amount_spend FROM overall GROUP BY name ORDER BY Amount_spend DESC LIMIT 5")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))

```

```python
                        #MONTHLY SALES TREND

mycursor.execute("SELECT EXTRACT(MONTH FROM order_date) AS Month, SUM(quantity * unit_price_usd) AS TotalSales FROM overall GROUP BY Month ORDER BY Month") 

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
                                  #SALES BY COUNTRY

mycursor.execute("SELECT country,SUM(quantity * unit_price_usd) AS TotalSales FROM overall GROUP BY country ORDER BY TotalSales DESC")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
                   #TOP 5 BEST SELLING PRODUCTS

mycursor.execute("SELECT product_name,SUM(quantity) AS Total_quantity_sold FROM overall GROUP BY product_name ORDER BY Total_quantity_sold DESC LIMIT 5")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
                #AVERAGE SALE PRICE BY PRODUCT

mycursor.execute("SELECT product_name, AVG(unit_price_usd) AS Average_Sale_Price FROM overall GROUP BY product_name")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))

```

```python
                       #CUSTOMER ACQUISITION OVER TIME

mycursor.execute("SELECT YEAR(order_date) AS Year, COUNT(customerkey) AS New_Customers FROM overall GROUP BY Year ORDER BY YEAR DESC")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
                             #AVERAGE ORDER BY MONTH

mycursor.execute("SELECT YEAR(order_date) AS Year, MONTH(order_date) AS Month, AVG(quantity) AS Average_order FROM overall GROUP BY Year,Month ORDER BY Year DESC,Month DESC")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))

```

```python
                         #TOP SELLING BRANDS

mycursor.execute("SELECT brand, SUM(quantity) AS Quantity_Sold FROM overall GROUP BY brand ORDER BY Quantity_Sold DESC")


out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description], tablefmt='psql'))

```
